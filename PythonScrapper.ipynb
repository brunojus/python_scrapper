```python
def request_until_succeed(url):
    # retrieve data
    http = urllib3.PoolManager()
    success = False
    while success is False:
        try: 
            req = http.request(method = 'GET', url = url)
            if req.status == 200:
                success = True
                response = req.data
        except e:
            print (e)
            time.sleep(15)
            
            print ("Error for URL %s: %s" % (url, datetime.datetime.now()))

    return response
```

```python
def getFacebookPageFeedData(page_id, access_token, num_statuses, num_comments):
    
    # construct the URL string
    base = "https://graph.facebook.com"
    node = "/" + page_id + "/feed"
    parameters = "/?fields=message,link,created_time,type,id,likes,comments.limit(%s){created_time,message,id,like_count}&limit=%s&access_token=%s" % (num_comments, num_statuses, access_token)
    # parameters = "/?fields=message,link,created_time,type,name,id,likes.limit(1).summary(true),comments.limit(1).summary(true),shares&limit=%s&access_token=%s" % (num_statuses, access_token) # changed
    url = base + node + parameters
    
    # retrieve data
    data = json.loads(request_until_succeed(url))
    
    return data
    

test_status = getFacebookPageFeedData(page_id, access_token, 1, 1)["data"][0]
print (json.dumps(test_status, indent=2, sort_keys=True))
```
